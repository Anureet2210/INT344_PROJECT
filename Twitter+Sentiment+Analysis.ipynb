```python
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
lemmatizer=WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    if tag.startswith('V'):
        return wordnet.VERB
    if tag.startswith('N'):
        return wordnet.NOUN
    if tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk.corpus import stopwords
import string
stop=stopwords.words('english')
stop+=list(string.punctuation)
stop
```

```python
def clean_data(words):
    output_words=[]
    for w in words:
        if w.lower() not in stop and not w.startswith('@'):
            pos=pos_tag(w)
            clean_words=lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            output_words.append(clean_words.lower())
    return output_words
```

```python
#fetching the clear training data
x_train_documents=[(clean_data(document))for document in x_train_data]
```
